```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

def load_ct_image(nifti_file_path):
    # Load the NIfTI file
    nii_img = nib.load(nifti_file_path)
    ct_image = nii_img.get_fdata()
    header = nii_img.header
    
    # Get voxel dimensions (spacing) from the header (in mm)
    voxel_dims = header.get_zooms()
    
    return ct_image, voxel_dims
```

```python
# Hounsfield Unit ranges for fat
FAT_HU_MIN = -190
FAT_HU_MAX = -30
```

```python
def segment_fat_tissue(ct_image):
    # Create a binary mask where fat tissue voxels are 1
    fat_mask = np.logical_and(ct_image >= FAT_HU_MIN, ct_image <= FAT_HU_MAX)
    return fat_mask.astype(np.uint8)


def calculate_volumes(fat_mask, voxel_dims):
    # Voxel volume in mm³
    voxel_volume = np.prod(voxel_dims)
    
    # Total abdominal volume
    total_volume = fat_mask.size * voxel_volume  # in mm³
    
    # Fat tissue volume
    fat_volume = np.sum(fat_mask) * voxel_volume  # in mm³
    
    return total_volume, fat_volume


def compute_fat_percentage(total_volume, fat_volume):
    fat_percentage = (fat_volume / total_volume) * 100
    return fat_percentage


def visualize_segmentation(ct_image, fat_mask, slice_index):
    plt.figure(figsize=(12, 6))

    # Original CT image slice
    plt.subplot(1, 2, 1)
    plt.imshow(ct_image[:, :, slice_index], cmap='gray')
    plt.title('Original CT Slice')
    plt.axis('off')

    # Fat tissue segmentation overlay
    plt.subplot(1, 2, 2)
    plt.imshow(ct_image[:, :, slice_index], cmap='gray')
    plt.imshow(fat_mask[:, :, slice_index], cmap='jet', alpha=0.5)
    plt.title('Fat Tissue Segmentation')
    plt.axis('off')

    plt.show()
```

## Fat-Height Histogram

```python
def plot_fat_distribution_along_height(fat_mask, voxel_dims):
    import matplotlib.pyplot as plt
    
    # Voxel dimensions (dx, dy, dz) in mm
    dx, dy, dz = voxel_dims
    
    # Calculate fat area per slice (in mm²)
    fat_area_per_slice = np.sum(fat_mask, axis=(0, 1)) * (dx * dy)
    
    # Calculate slice positions along the z-axis (in mm)
    num_slices = fat_mask.shape[2]
    slice_positions = np.arange(num_slices) * dz  # Starting from 0 mm
    
    # Plot the fat area along the height
    plt.figure(figsize=(12, 6))
    plt.plot(slice_positions, fat_area_per_slice, color='red', linewidth=2)
    plt.xlabel('Height along Z-axis (mm)')
    plt.ylabel('Fat Area per Slice (mm²)')
    plt.title('Fat Distribution Along the Person\'s Height')
    plt.grid(True)
    plt.show()


def plot_fat_heatmap(fat_mask, voxel_dims):
    import matplotlib.pyplot as plt
    
    # Sum fat along the x-axis to get a 2D map (y vs. z)
    fat_distribution = np.sum(fat_mask, axis=0) * voxel_dims[0]  # Sum over x-axis and multiply by voxel width
    
    plt.figure(figsize=(8, 10))
    plt.imshow(fat_distribution.T, cmap='hot', aspect='auto', origin='lower')
    plt.colorbar(label='Fat Thickness (mm)')
    plt.xlabel('Y-axis Pixels')
    plt.ylabel('Slice Number (Z-axis)')
    plt.title('Heatmap of Fat Distribution (Y vs. Height)')
    plt.show()
```

```python
def abdominal_fat_percentage_poc(nifti_file_path):
    # Load CT image
    ct_image, voxel_dims = load_ct_image(nifti_file_path)
    
    # Segment fat tissue
    fat_mask = segment_fat_tissue(ct_image)
    
    # Calculate volumes
    total_volume, fat_volume = calculate_volumes(fat_mask, voxel_dims)
    
    # Compute fat percentage
    fat_percentage = compute_fat_percentage(total_volume, fat_volume)
    
    print(f"Abdominal Fat Percentage: {fat_percentage:.2f}%")
    
    # Visualize (optional)
    # Choose a slice index to visualize (e.g., middle slice)
    slice_index = ct_image.shape[2] // 2
    visualize_segmentation(ct_image, fat_mask, slice_index)
    
    # Plot fat distribution along height
    plot_fat_distribution_along_height(fat_mask, voxel_dims)
    plot_fat_heatmap(fat_mask, voxel_dims)
```

```python
def create_slice_animation(ct_image, fat_mask):
    import matplotlib.animation as animation

    fig, ax = plt.subplots(figsize=(6, 6))
    fig.suptitle('Fat Segmentation Animation')

    slice_index = 0
    img = ax.imshow(ct_image[:, :, slice_index], cmap='gray')
    seg = ax.imshow(fat_mask[:, :, slice_index], cmap='jet', alpha=0.5)
    ax.axis('off')

    def update(frame):
        img.set_data(ct_image[:, :, frame])
        seg.set_data(fat_mask[:, :, frame])
        return [img, seg]

    ani = animation.FuncAnimation(fig, update, frames=ct_image.shape[2], interval=50, blit=True)
    plt.show()
```

```python
def plot_cumulative_fat_volume(fat_mask, voxel_dims):
    import matplotlib.pyplot as plt
    
    # Voxel volume in mm³
    voxel_volume = np.prod(voxel_dims)
    
    # Calculate fat volume per slice
    fat_volume_per_slice = np.sum(fat_mask, axis=(0, 1)) * voxel_volume
    
    # Cumulative sum
    cumulative_fat_volume = np.cumsum(fat_volume_per_slice)
    
    # Slice positions along z-axis
    num_slices = fat_mask.shape[2]
    slice_positions = np.arange(num_slices) * voxel_dims[2]
    
    # Plot cumulative fat volume
    plt.figure(figsize=(12, 6))
    plt.plot(slice_positions, cumulative_fat_volume, color='green', linewidth=2)
    plt.xlabel('Height along Z-axis (mm)')
    plt.ylabel('Cumulative Fat Volume (mm³)')
    plt.title('Cumulative Fat Volume Along the Person\'s Height')
    plt.grid(True)
    plt.show()
```

```python
def interactive_segmentation_viewer(ct_image):
    from ipywidgets import interact, FloatSlider
    import matplotlib.pyplot as plt
    
    def view_slice(slice_index, hu_min, hu_max):
        slice_data = ct_image[:, :, int(slice_index)]
        fat_mask = np.logical_and(slice_data >= hu_min, slice_data <= hu_max)
        
        plt.figure(figsize=(8, 8))
        plt.imshow(slice_data, cmap='gray')
        plt.imshow(fat_mask, cmap='jet', alpha=0.5)
        plt.title(f'Slice {int(slice_index)} with HU [{hu_min}, {hu_max}]')
        plt.axis('off')
        plt.show()
    
    interact(view_slice,
             slice_index=(0, ct_image.shape[2]-1, 1),
             hu_min=FloatSlider(value=-190, min=-300, max=0, step=1),
             hu_max=FloatSlider(value=-30, min=-300, max=100, step=1))
```

## Estimate full-body fat percentage using heuristics

```python
def estimate_full_body_fat_percentage(fat_mask, ct_image, voxel_dims, patient_height=1750):
    """
    Estimates the full-body fat percentage using abdominal CT data and heuristics.
    
    Parameters:
    - fat_mask: Binary mask of fat tissue (3D numpy array).
    - ct_image: Original CT image data (3D numpy array).
    - voxel_dims: Tuple of voxel dimensions (dx, dy, dz) in mm.
    - patient_height: Height of the patient in mm (default 1750 mm for 175 cm).
    
    Returns:
    - body_fat_percentage: Estimated full-body fat percentage.
    """
    import numpy as np
    
    # Voxel dimensions
    dx, dy, dz = voxel_dims
    
    # Calculate Abdominal Fat Volume (in mm³)
    voxel_volume = dx * dy * dz  # Voxel volume in mm³
    abdominal_fat_volume = np.sum(fat_mask) * voxel_volume  # Total fat volume in abdomen
    
    # Estimate Total Fat Volume
    percentage_abdominal_fat = 0.55  # Assume 55% of total fat is in the abdomen for males
    total_fat_volume = abdominal_fat_volume / percentage_abdominal_fat
    
    # Estimate Total Body Volume
    # Step 1: Calculate average abdominal cross-sectional area
    abdominal_area_per_slice = np.sum(ct_image > -1024, axis=(0, 1)) * (dx * dy)  # Exclude air (-1024 HU)
    average_abdominal_area = np.mean(abdominal_area_per_slice)  # in mm²
    
    # Step 2: Estimate body radius from abdominal area
    average_body_radius = np.sqrt(average_abdominal_area / np.pi)  # in mm
    
    # Step 3: Estimate total body volume (cylinder model)
    total_body_volume = np.pi * (average_body_radius ** 2) * patient_height  # in mm³
    
    # Calculate Body Fat Percentage
    body_fat_percentage = (total_fat_volume / total_body_volume) * 100  # In percentage
    
    return body_fat_percentage
```
